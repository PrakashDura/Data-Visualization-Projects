## Part 1 -- Data and Visualization


```python
#include all the modules here
from scipy.stats import linregress
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
```

```python
# Import the Iris dataset:
iris_dataset = sns.load_dataset('iris')
# Show Iris data to check if the data is being uploaded correctly
iris_dataset.head()
```

```python
#Create a Pandas Dataframe to store the data
df = pd.DataFrame(iris_dataset)
print(df)
```

```python
# Show samples from each species
iris_dataset["species"].value_counts()
```

```python
# Create a dataframe for only setosa and versicolor
newdata = iris_dataset[(iris_dataset['species'] == 'setosa') | (iris_dataset['species'] == 'versicolor')]

# display the new dataframe
print(newdata)

```

```python
# Show pairwise relationships in this dataset using seaborn's pairplot 
sns.pairplot(newdata, hue='species')
plt.show()
```

```python
# draw a subplot for two figures such as:
# 1) petal_length vs. petal_width
# 2) petal_length and sepal_width
fig = plt.figure(figsize=(10, 5))

# subplot 1: petal_length vs. petal_width
plt.subplot(1, 2, 1)
sns.scatterplot(data=newdata, x='petal_length', y='petal_width', hue='species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs. Petal Width')
plt.legend()

# subplot 2: petal_length vs. sepal_width
plt.subplot(1, 2, 2)
sns.scatterplot(data=newdata, x='petal_length', y='sepal_width', hue='species')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Width')
plt.title('Petal Length vs. Sepal Width')
plt.legend()
plt.tight_layout()
plt.show()


```

```python
# find the means of each specimens in each graph and calculate the distance between the means and draw a line for each subplot
# with a text showing the distance use plt.text
#finding the mean of setosa and versicolor

setosa_means = newdata.loc[newdata['species'] == 'setosa', ['petal_length', 'petal_width', 'sepal_width']].mean()
versicolor_means = newdata.loc[newdata['species'] == 'versicolor', ['petal_length', 'petal_width', 'sepal_width']].mean()

# calculating distances
petal_length_width_dist = np.sqrt(np.sum((setosa_means[['petal_length', 'petal_width']] - versicolor_means[['petal_length', 'petal_width']])**2))
petal_length_sepal_width_dist = np.sqrt(np.sum((setosa_means[['petal_length', 'sepal_width']] - versicolor_means[['petal_length', 'sepal_width']])**2))

# print distances
print("Distance between setosa and versicolor means for 'petal_length' and 'petal_width':", petal_length_width_dist)
print("Distance between setosa and versicolor means for 'petal_length' and 'sepal_width':", petal_length_sepal_width_dist)

# plot a sub plot of petal length vs. petal width and petal length vs. sepal width

# create subplots
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=newdata, x='petal_length', y='petal_width', hue='species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs. Petal Width for setosa & versicolor')

# plot line and text for petal_length vs. petal_width
plt.plot([setosa_means['petal_length'], versicolor_means['petal_length']],
         [setosa_means['petal_width'], versicolor_means['petal_width']], color='black', marker='.', )
plt.text((setosa_means['petal_length'] + versicolor_means['petal_length'])/2,
         (setosa_means['petal_width'] + versicolor_means['petal_width'])/2,
        "2.998056036834535", ha='center')

#create subplots
plt.subplot(1, 2, 2)
sns.scatterplot(data=newdata, x='petal_length', y='sepal_width', hue='species')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Width')
plt.title('Petal Length vs. Sepal Width for setosa & versicolor')

# plot line and text for petal_length vs. sepal_width
plt.plot([setosa_means['petal_length'], versicolor_means['petal_length']],
         [setosa_means['sepal_width'], versicolor_means['sepal_width']], color='black',marker='.')
plt.text((setosa_means['petal_length'] + versicolor_means['petal_length'])/2,
         (setosa_means['sepal_width'] + versicolor_means['sepal_width'])/2,
         "2.8701080119047786", ha='center')

plt.tight_layout()
plt.show()
```

```python
# This time use seaborn's jointplot to show the regression line

# creating a jointplot for petal_length vs. petal_width
sns.jointplot(data=newdata, x='petal_length', y='petal_width', kind= 'reg')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.suptitle('Petal Length vs. Petal Width for setosa & versicolor')

# creating a jointplot for petal_length vs. sepal_width
sns.jointplot(data=newdata, x='petal_length', y='sepal_width',  kind='reg')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Width')
plt.suptitle('Petal Length vs. Sepal Width for setosa & versicolor')



# display the jointplots
plt.show()
```

```python
# Use the given regression code in the class to find the regression line for only setosa for petal_length and petal_width
# filter the iris dataset to include only "setosa" species
setosa_data = iris_dataset[iris_dataset['species'] == 'setosa']

# fit a linear regression model to the setosa petal_length vs. petal_width data
slope, intercept, r_value, p_value, std_err = linregress(setosa_data['petal_length'], setosa_data['petal_width'])

# plot the scatter plot of the setosa petal_length vs. petal_width data and overlay the regression line
sns.scatterplot(data=setosa_data, x='petal_length', y='petal_width')
plt.plot(setosa_data['petal_length'], intercept + slope * setosa_data['petal_length'], color='red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Regression of Setosa Petal Length vs. Width')

# display the plot
plt.show()
```

## Part 3 -- Correlation between features 


```python
# This time use all the data for all the species (setosa, versicolor, virginica)
# Use the seaborn correlation matrix and heatmaps to visualize)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# load the iris dataset into a pandas dataframe
iris_dataset = sns.load_dataset('iris')

```

```python
# compute the correlation matrix
corr_matrix = iris_dataset.corr()

```

```python
# create a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, cmap= 'coolwarm', annot=True)

# display the heatmap
plt.show()

# find the two features that correlate the most
max_corr = corr_matrix.stack().sort_values(ascending=False).reset_index()
max_corr.columns = ['feature1', 'feature2', 'correlation']
max_corr = max_corr[max_corr['correlation'] != 1]
print("The two features that correlate the most are:")
print(max_corr.iloc[0]['feature1'], "and", max_corr.iloc[0]['feature2'])
```
